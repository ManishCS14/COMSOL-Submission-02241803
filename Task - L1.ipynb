```python
#Step 1: Import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
```

```python
#Step 2: Plot the first function
x = np.linspace(0,np.pi,100)
y = np.sin(x)
a_range = [np.pi/2, np.pi/3, np.pi/4]
x_1 = [0, np.pi/2, np.pi]
x_2 = [0, np.pi/3, (2*np.pi)/3, np.pi]
x_3 = [0, np.pi/4, np.pi/2, (3*np.pi)/4, np.pi]
y_1 = np.sin(x_1)
y_2 = np.sin(x_2)
y_3 = np.sin(x_3)
x_interp = np.pi / 4
y_interp = np.interp(x_interp, x_1, y_1)
plt.plot(x,y, c ='b', label = 'Exact Function')
plt.plot(x_1,y_1, c = 'r', ls = '--', label = 'Interpolated Function')
plt.xticks([0,np.pi/4,np.pi/2,(3*np.pi)/4, np.pi], ['0',r'$π/4$',r'$π/2$',r'$3π/4$',r'$π$'])
x_points = [np.pi/4, 3*np.pi/4]
for xp in x_points:
    plt.plot([xp, xp], [0, np.sin(xp)], 'k--', linewidth=1.2)  # vertical line
    plt.plot([0, xp], [np.sin(xp), np.sin(xp)], 'k--', linewidth=1.2)  # horizontal line

# Add custom ticks at specific points with LaTeX notation
plt.xticks([0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi], [r'$0$', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$', r'$\pi$'])
plt.xlim([0,np.pi])
plt.ylim([0,1.2])
# Annotate the points at pi/4
exact_value_at_pi4 = np.sin(np.pi/4)
plt.annotate(f'Exact: {exact_value_at_pi4:.4f}', 
             xy=(np.pi/4, exact_value_at_pi4), 
             xytext=(np.pi/4 + 0.1, exact_value_at_pi4 + 0.05),
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10)
plt.annotate(f'Interpolated: {y_interp:.4f}', 
             xy=(np.pi/4, y_interp), 
             xytext=(np.pi/4 + 0.1, y_interp - 0.1),
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10)
plt.title('Sine Function and Interpolated Function with Mesh Spacing π/2')
plt.xlabel(r'$\theta$ (radians)')
plt.ylabel(r'$\sin(\theta)$')
plt.legend(loc = 'upper right', frameon = False)
plt.tight_layout()
plt.show()
```

```python
# Set up the range and the sine function
x = np.linspace(0, np.pi, 100)
y = np.sin(x)

# Mesh spacing π/3
x_2 = [0, np.pi/3, (2*np.pi)/3, np.pi]  # Interpolation points
y_2 = np.sin(x_2)

# Interpolation to find the value at π/6
x_interp = np.pi / 6
y_interp = np.interp(x_interp, x_2, y_2)

# Plot the exact function and the interpolated function
plt.plot(x, y, c='b', label='Exact Function')
plt.plot(x_2, y_2, c='r', ls='--', label='Interpolated Function')

# Set ticks for x-axis
plt.xticks([0, np.pi/6, np.pi/3, np.pi/2, (2*np.pi)/3, 5*np.pi/6, np.pi],
           [r'$0$', r'$\frac{\pi}{6}$', r'$\frac{\pi}{3}$', r'$\frac{\pi}{2}$', r'$\frac{2\pi}{3}$', r'$\frac{5\pi}{6}$', r'$\pi$'])

# Add vertical and horizontal lines at specific points (π/6 and π/2)
x_points = [np.pi/6, np.pi/2]
for xp in x_points:
    plt.plot([xp, xp], [0, np.sin(xp)], 'k--', linewidth=1.2)  # vertical line
    plt.plot([0, xp], [np.sin(xp), np.sin(xp)], 'k--', linewidth=1.2)  # horizontal line

# Set plot limits
plt.xlim([0, np.pi])
plt.ylim([0, 1.2])

# Annotate the points at π/6 (exact and interpolated)
exact_value_at_pi6 = np.sin(np.pi/6)
plt.annotate(f'Exact: {exact_value_at_pi6:.4f}', 
             xy=(np.pi/6, exact_value_at_pi6), 
             xytext=(np.pi/6 + 0.1, exact_value_at_pi6 + 0.05),
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10)
plt.annotate(f'Interpolated: {y_interp:.4f}', 
             xy=(np.pi/6, y_interp), 
             xytext=(np.pi/6 + 0.1, y_interp - 0.1),
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10)

# Annotate the points at π/2 (exact and interpolated) - moved to top left to avoid clutter
exact_value_at_pi2 = np.sin(np.pi/2)
interp_value_at_pi2 = np.interp(np.pi/2, x_2, y_2)
plt.annotate(f'Exact: {exact_value_at_pi2:.4f}', 
             xy=(np.pi/2, exact_value_at_pi2), 
             xytext=(np.pi/2 - 0.9, exact_value_at_pi2 + 0.05),  # shifted left and up
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10)
plt.annotate(f'Interpolated: {interp_value_at_pi2:.4f}', 
             xy=(np.pi/2, interp_value_at_pi2), 
             xytext=(np.pi/2 + 0.1, interp_value_at_pi2 - 0.1),
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10)

# Add title, labels, and legend
plt.title('Sine Function and Interpolated Function with Mesh Spacing π/3')
plt.xlabel(r'$\theta$ (radians)')
plt.ylabel(r'$\sin(\theta)$')
plt.legend(loc='upper right', frameon=False)

# Ensure the plot layout is neat
plt.tight_layout()
plt.show()
```

```python
# Set up the range and the sine function
x = np.linspace(0, np.pi, 100)
y = np.sin(x)

# Mesh spacing π/4
x_2 = [0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi]  # Interpolation points
y_2 = np.sin(x_2)

# Interpolation to find the value at π/8 and 3π/8
x_interp1 = np.pi / 8
y_interp1 = np.interp(x_interp1, x_2, y_2)

x_interp2 = 3 * np.pi / 8
y_interp2 = np.interp(x_interp2, x_2, y_2)

# Plot the exact function and the interpolated function
plt.plot(x, y, c='b', label='Exact Function')
plt.plot(x_2, y_2, c='r', ls='--', label='Interpolated Function')

# Set ticks for x-axis
plt.xticks([0, np.pi/8, np.pi/4, 3*np.pi/8, np.pi/2, 5*np.pi/8, 3*np.pi/4, 7*np.pi/8, np.pi],
           [r'$0$', r'$\frac{\pi}{8}$', r'$\frac{\pi}{4}$', r'$\frac{3\pi}{8}$', r'$\frac{\pi}{2}$',
            r'$\frac{5\pi}{8}$', r'$\frac{3\pi}{4}$', r'$\frac{7\pi}{8}$', r'$\pi$'])

# Add vertical and horizontal lines at specific points (π/8 and 3π/8)
x_points = [np.pi/8, 3*np.pi/8]
for xp in x_points:
    plt.plot([xp, xp], [0, np.sin(xp)], 'k--', linewidth=1.2)  # vertical line
    plt.plot([0, xp], [np.sin(xp), np.sin(xp)], 'k--', linewidth=1.2)  # horizontal line

# Set plot limits
plt.xlim([0, np.pi])
plt.ylim([0, 1.2])

# Annotate the points at π/8 (exact and interpolated)
exact_value_at_pi8 = np.sin(np.pi/8)
plt.annotate(f'Exact: {exact_value_at_pi8:.4f}', 
             xy=(np.pi/8, exact_value_at_pi8), 
             xytext=(np.pi/8 + 0.1, exact_value_at_pi8 + 0.05),
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10)
plt.annotate(f'Interpolated: {y_interp1:.4f}', 
             xy=(np.pi/8, y_interp1), 
             xytext=(np.pi/8 + 0.1, y_interp1 - 0.1),
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10)

# Annotate the points at 3π/8 (exact and interpolated)
exact_value_at_3pi8 = np.sin(3*np.pi/8)
plt.annotate(f'Exact: {exact_value_at_3pi8:.4f}', 
             xy=(3*np.pi/8, exact_value_at_3pi8), 
             xytext=(3*np.pi/8 - 0.4, exact_value_at_3pi8 + 0.1),  # Top left annotation for exact value
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10)
plt.annotate(f'Interpolated: {y_interp2:.4f}', 
             xy=(3*np.pi/8, y_interp2), 
             xytext=(3*np.pi/8 + 0.2, y_interp2 - 0.15),  # Lower mid-right annotation for interpolated value
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10)

# Add title, labels, and legend
plt.title('Sine Function and Interpolated Function with Mesh Spacing π/4')
plt.xlabel(r'$\theta$ (radians)')
plt.ylabel(r'$\sin(\theta)$')
plt.legend(loc='upper right', frameon=False)

# Ensure the plot layout is neat
plt.tight_layout()
plt.show()
```
