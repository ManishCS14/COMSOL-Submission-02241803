```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Data points
potential = np.array([0, 4, 8, 12, 16, 20])
deflection = np.array([0, 16.566, 33.132, 49.698, 66.264, 82.830])

# Reshape for sklearn linear regression
X = potential.reshape(-1, 1)
y = deflection

# Linear regression model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Calculate slope, intercept, and R^2 value
slope = model.coef_[0]
intercept = model.intercept_
r_squared = r2_score(y, y_pred)

# Create plot
plt.figure(figsize=(8, 6))
plt.plot(potential, deflection, 'ro', label="Data Points")  # Red markers for data points
plt.plot(potential, y_pred, 'r-', linewidth=1.5)  # Slightly thinner red line for the fit

# Equation and R^2 text on plot in scientific notation for intercept
equation_text = f"y = {slope:.4f}x + {intercept:.2e}\n$R^2$ = {r_squared:.4f}"
plt.text(0.05, 0.95, equation_text, transform=plt.gca().transAxes, 
         verticalalignment='top', horizontalalignment='left', fontsize=12)

# Labels
plt.xlabel("Applied Voltage (V)", fontsize=12, fontweight='bold')
plt.ylabel("Deflection (nm)", fontsize=12, fontweight='bold')

# Show plot
plt.tight_layout()
plt.show()
```
